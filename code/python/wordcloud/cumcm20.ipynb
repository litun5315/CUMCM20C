# 数据处理-提取公司类型关键词
- 输入：tb_idx - 表格附件编号（1和2）
- 处理过程：产生关键词类型关键词列添加在第三列，并生成一个新表格

```python
import xlrd

tb_idx = 2
pth = '/Users/teng/Desktop/cumcm20/数据/处理/'
tb_pth = pth + str(tb_idx) + '.xlsx'
tb_wk = xlrd.open_workbook(tb_pth)
```

```python
tb_st1 = tb_wk.sheets()[0]
tb_data = []
for i in range(1, tb_st1.nrows):
    temp = []
    for j in range(2):
        temp.append(tb_st1.cell(i, j).value)
    tb_data.append(temp)
    
# for i in range(tb_st1.nrows - 1):
#     for j in range(2):
#         print(tb_data[i][j], end = '\t')
#     print(' ')
```

```python
import re

for i in range(len(tb_data)):
    temp = tb_data[i][1]
    temp = re.sub(r'^[*][*][*]', '', temp)
    temp = re.sub(r'E\d* $', '', temp)
    temp = re.sub(r'[*][*][*]分公司$', '', temp)
    temp = re.sub(r'[*][*][*]分院$', '', temp)
    temp = re.sub(r'[*][*][*]药店$', '', temp)
    temp = re.sub(r'有限公司$', '', temp)
    temp = re.sub(r'有限责任公司$', '', temp)
    temp  = re.sub(r'（中国）$', '', temp)
    temp  = re.sub(r'（有限合伙）$', '', temp)
    
    tb_data[i].append(temp)
```

```python
for i in range(len(tb_data)):
    print(tb_data[i])
```

```python
import xlwt

workbook = xlwt.Workbook(encoding = 'utf-8')
worksheet = workbook.add_sheet('公司种类')
for i in range(len(tb_data)):
    worksheet.write(i+1,0, label = tb_data[i][0])
    worksheet.write(i+1,1, label = tb_data[i][1])
    worksheet.write(i+1,2, label = tb_data[i][2])
save_pth = pth + str(tb_idx) + '_coptype.xls'
workbook.save(save_pth)
```
